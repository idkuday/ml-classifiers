```python
df = pd.read_csv("./classification/linearlySeparable/class1_val.txt", header =None,delimiter = ' ')
c1_val = np.array(df.values.tolist())
print(c1_val.shape)
plt.figure()
plt.scatter(c1_train[:,0],c1_train[:,1],color='red')
plt.scatter(c2_train[:,0],c2_train[:,1],color='green')
plt.scatter(c3_train[:,0],c3_train[:,1],color='blue')
plt.scatter(np.mean(c2_train[:,0]),np.mean(c2_train[:,1]),color ='k')
plt.scatter(np.mean(c1_train[:,0]),np.mean(c1_train[:,1]),color ='k')
plt.scatter(np.mean(c3_train[:,0]),np.mean(c3_train[:,1]),color ='k')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linearly Separable Training data and its Mean")
plt.show()
```

```python
dist = list()
for i in c1_train :
    dist.append([np.linalg.norm(c1_val[0]-i),'1'])

for i in c2_train :
    dist.append([np.linalg.norm(c1_val[0]-i),'2'])
for i in c3_train :
    dist.append([np.linalg.norm(c1_val[0]-i),'3'])
dist.sort(key = lambda dist:dist[0]) #lambda for sorting criteria
print(dist)

```
